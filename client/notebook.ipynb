```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/mattiesanseverino/Downloads/landfills_xy.csv")
info = pd.read_csv("/Users/mattiesanseverino/Downloads/landfill_info.csv")
```

```python
grouped_df = df.groupby('SWIS').agg({'Longitude (Decimal Degrees)': list, 'Latitude (Decimal Degrees)': list}).reset_index()
```

```python
joined_df = pd.merge(grouped_df, info, on='SWIS', how='outer')
joined_df.head()
```

```python
#joined_df['SWIS'] = grouped_df['SWIS'].apply(lambda x: f'"{x}"')
#joined_df['Longitude (Decimal Degrees)'] = grouped_df['Longitude (Decimal Degrees)'].apply(lambda x: f'"{x}"')
#joined_df['Latitude (Decimal Degrees)'] = grouped_df['Latitude (Decimal Degrees)'].apply(lambda x: f'"{x}"')
for col in joined_df.columns:
    joined_df[col] = joined_df[col].apply(lambda x: f'"{x}"')
final_df = joined_df[['SWIS', 'Longitude (Decimal Degrees)', 'Latitude (Decimal Degrees)', 'Landfill Name', 'City', 'Landfill Ownership Type', 'Owner', 'Class', 'Type of Waste', 'Type of Waste 2 (Supporting Info)', 'Current Landfill Status', 'Year Opened', 'Year Closed', 'Year Range if not Exact', 'Converted', 'Conversion Type', 'Site Conversion ', 'Site Regulartory Status', 'Documentation Link']]
```
